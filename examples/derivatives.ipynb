```python
import keyring
from pycgapi import CoinGeckoAPI
```

```python
api_key = keyring.get_password('coingecko', 'analyst_api_key')
```

```python
api = CoinGeckoAPI(api_key, pro_api=True)
status = api.status_check()
print("API Server Status:", status)
```

## /derivatives

```python
derivatives_tickers = api.derivatives_market_tickers()
derivatives_tickers
```

## /derivatives/exchanges

```python
derivatives_exchanges = api.derivatives_exchanges_list()
derivatives_exchanges.head(10)
```

```python
exchange_id = 'binance_futures'
```

## /derivatives/exchanges/{id}

```python
derivatives_exchange_data = api.derivatives_exchange_info(exchange_id)
derivatives_exchange_data
```
