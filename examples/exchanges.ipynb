```python
import time
import keyring
from coingeckoapi import CoinGeckoAPI
```

```python
api_key = keyring.get_password('coingecko', 'analyst_api_key')
```

```python
api = CoinGeckoAPI(api_key, pro_api=True)
status = api.status_check()
print("API Server Status:", status)
```

## /exchanges

```python
all_exchanges = api.active_exchanges_list()
all_exchanges.head()
```

## /exchanges/list

```python
supported_exchanges = api.all_exchanges_list()
supported_exchanges
```

```python
exchange_id='gdax'
```

## /exchanges/{id}

```python
exchange_volume = api.exchange_volume_data(exchange_id)
print('Available Keys: {}'.format(sorted(exchange_volume.keys())))
```

## /exchanges/{id}/tickers

```python
exchange_tickers = api.exchange_market_tickers(exchange_id)
exchange_tickers.head()
```

## /exchanges/{id}/volume_chart

```python
exchange_volume_data = api.exchange_historical_volume(exchange_id)
exchange_volume_data
```

## /exchanges/{id}/volume_chart/range

```python
exchange_volume_between_dates = api.exchange_historical_volume(
    exchange_id, 
    from_date='01-01-2023',
    to_date='01-31-2023'
)
exchange_volume_between_dates
```
